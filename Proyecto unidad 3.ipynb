```python
# Definir una funcin para mapear los valores de "S" a 1 y "No" a 0
def map_si_no(valor):
    if valor == "S":
        return 1
    elif valor == "No":
        return 0
    else:
        return None  # Manejar cualquier otro valor (puede ser NaN)

# Crear las nuevas columnas DIABETES_NUM y OBESIDAD_NUM aplicando la funcin map_si_no
df['DIABETES_NUM'] = df['DIABETES'].apply(lambda x: map_si_no(x))
df['OBESIDAD_NUM'] = df['OBESIDAD'].apply(lambda x: map_si_no(x))
df['UCI_NUM'] = df['UCI'].apply(lambda x: map_si_no(x))
df['TABAQUISMO_NUM'] = df['TABAQUISMO'].apply(lambda x: map_si_no(x))
df['HIPERTENSION_NUM'] = df['HIPERTENSION'].apply(lambda x: map_si_no(x))
df['CARDIOVASCULAR_NUM'] = df['CARDIOVASCULAR'].apply(lambda x: map_si_no(x))
df['ASMA_NUM'] = df['ASMA'].apply(lambda x: map_si_no(x))
df['INMUSUPR_NUM'] = df['INMUSUPR'].apply(lambda x: map_si_no(x))
df['EPOC_NUM'] = df['EPOC'].apply(lambda x: map_si_no(x))
df['RENAL_CRONICA_NUM'] = df['RENAL_CRONICA'].apply(lambda x: map_si_no(x))
```

## 4. Modelo de rbol de Decisin
###  Entrenar un modelo de rbol de Decisin utilizando las caractersticas seleccionadas.
###  Evaluar el desempeo del modelo utilizando mtricas como el coeficiente de determinacin (R) y la matriz de confusin.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Crear una instancia del modelo de rbol de Decisin
modelo_arbol = DecisionTreeClassifier()
```
